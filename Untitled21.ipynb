```python
# Install required libraries
!pip install transformers gradio

# Import necessary libraries
import json
from transformers import pipeline
import gradio as gr

# Load the JSON data
file_path = 'Campus_Pal.json'  # Adjust based on the file's path in Colab
with open(file_path, 'r') as file:
    data = json.load(file)  # Assume data is a list of dictionaries

# Preprocess the data into a dictionary for faster lookup
context_dict = {}
for item in data:  # Iterate over the list
    for key, value in item.items():  # Access dictionary key-value pairs
        context_dict[key] = value

# Load the pre-trained QA model
qa_model = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Define a function to filter relevant context
def find_relevant_context(question):
    keywords = question.lower().split()  # Split question into words
    relevant_context = ""
    for key, value in context_dict.items():
        if any(keyword in key.lower() or keyword in str(value).lower() for keyword in keywords):
            relevant_context += f"{key}: {value}\n"
    return relevant_context if relevant_context else "Context not found for the question."

# Define a function for answering questions
def ask_question(question):
    relevant_context = find_relevant_context(question)
    if relevant_context == "Context not found for the question.":
        return relevant_context
    result = qa_model(question=question, context=relevant_context)
    return result['answer']

# Create a Gradio interface
def qna_bot(question):
    return ask_question(question)

# Launch the app
interface = gr.Interface(
    fn=qna_bot,
    inputs="text",
    outputs="text",
    title="Campus Pal Q&A Bot",
    description="Ask any questions related to the Campus Pal dataset. Powered by a better-performing model for accurate responses.",
)

# Run the app
interface.launch(share=True)
```
